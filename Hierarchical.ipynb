```python
# Importing necessary libraries
from sklearn.datasets import make_blobs #make_blobs generates a simple dataset
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Generate sample data
X, _ = make_blobs(n_samples=100, centers=2, random_state=42, cluster_std=1.0)

# Plot the dataset
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Sample Data")
plt.show()

# Take user input for number of clusters and linkage method
n_clusters = int(input("Enter the number of clusters: "))
linkage_method = input("Enter the linkage method (complete, average, single): ")

# Initialize Agglomerative Clustering model
agg_clustering = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage_method)

# Fit the model
agg_clustering.fit(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=agg_clustering.labels_, cmap='rainbow')
plt.show()
```

make_blobs : It creates clusters of data points for testing clustering algorithms.
AgglomerativeClustering: This is the class used to perform hierarchical clustering
