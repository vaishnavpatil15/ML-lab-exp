```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate synthetic data for clustering
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Create a DBSCAN object with epsilon=0.5 and min_samples=5
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Fit the DBSCAN model to the data
dbscan.fit(X)

# Get the cluster labels
labels = dbscan.labels_

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

eps (epsilon) is the maximum distance between two samples for them to be considered as in the same neighborhood.
min_samples is the minimum number of samples in a neighborhood for a point to be considered a core point. This parameter helps to define the density of a cluster.
